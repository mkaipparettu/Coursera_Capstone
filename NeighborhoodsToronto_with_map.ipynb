```python
df2= data[0]
df2.head(10)
```

####  Ignore cells with a Borough that is Not assigned

```python
df2 = df2[df2.Borough != 'Not assigned']
df2.head(10)
```

#### If  Neighbourhood is  Not assigned then replace it with Borough

```python
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned', df2['Borough'], df2['Neighbourhood'])
df2.head(10)
```

```python
df2 = df2.sort_values(by=['Postcode', 'Borough']).groupby('Postcode').aggregate({'Borough':'first','Neighbourhood': ', '.join}).reset_index()
df2.head(10)
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="t_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_t = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_t)  
    
map_t
```
